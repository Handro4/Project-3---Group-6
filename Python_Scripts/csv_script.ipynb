```python
# Import dependencies
import pandas as pd
import csv as CsV
import json
import os
```

```python
# Read in CSV
file = pd.read_csv("../U.S._Chronic_Disease_Indicators__CDI_.csv")
csv = pd.DataFrame(file)
```

```python
#Preview the CSV
csv.head()
```

```python
# Filter down to Topics we want to look at
cdi_csv = csv[(csv["Topic"] == "Asthma") | (csv["Topic"] == "Chronic Obstructive Pulmonary Disease") | (csv["Topic"] == "Cardiovascular Disease")]
```

```python
# Filter down by year
cdi_csv_year = cdi_csv[(cdi_csv["YearStart"] == 2020)]
```

```python
# Preview the data to see the data types
cdi_csv_year.dtypes
```

```python
# Change the dtype of the Value so we can take the mean
cdi_csv_year["DataValue"] = cdi_csv_year["DataValue"].astype(float)
```

```python
# Create new columns to separate lat and long numbers
cdi_csv_year["GeoLocation"] = cdi_csv_year['GeoLocation'].str.replace('POINT ', '').str.replace('(', '').str.replace(')', '')
cdi_csv_year[["Latitude", "Longitude"]] = cdi_csv_year['GeoLocation'].str.split(' ', n=1, expand=True)
cdi_csv_clean = cdi_csv_year.drop(["DataSource", "GeoLocation", "Response", "DataValueUnit", "DataValueFootnoteSymbol", "DatavalueFootnote", "StratificationCategory2",	"Stratification2",	"StratificationCategory3",	"Stratification3", "ResponseID", "TopicID",	"QuestionID",	"DataValueTypeID",	"StratificationCategoryID1",	"StratificationID1",	"StratificationCategoryID2",	"StratificationID2",	"StratificationCategoryID3",	"StratificationID3"], axis=1)
cdi_csv_clean["Latitude"] = cdi_csv_clean["Latitude"].astype(float)
cdi_csv_clean["Longitude"] = cdi_csv_clean["Longitude"].astype(float)
```

```python
# Check the dtypes
cdi_csv_clean.dtypes
```

```python
# Group by Topic and Question and then average
cdi_csv_group = cdi_csv_clean.groupby(["Topic", "Question"])["DataValue"].mean().round(2)
cdi_csv_group
```

```python
# Creating an asthma csv to hold our data
asthma_csv = cdi_csv_clean[(cdi_csv_clean["Topic"] == "Asthma") & (cdi_csv_clean["Question"] == "Current asthma prevalence among adults aged >= 18 years")]

```

```python
# Cardiovascular CSV
cardiovascular_csv = cdi_csv_clean[(cdi_csv_clean["Topic"] == "Cardiovascular Disease") & (cdi_csv_clean["Question"] == "Mortality from total cardiovascular diseases")]
```

```python
# COPD CSV
copd_csv = cdi_csv_clean[(cdi_csv_clean["Topic"] == "Chronic Obstructive Pulmonary Disease") & (cdi_csv_clean["Question"] == "Prevalence of chronic obstructive pulmonary disease among adults >= 18")]
```

```python
# Merge Data into a single dataframe
final_csv = pd.concat([asthma_csv, cardiovascular_csv, copd_csv], ignore_index=True)
final_csv.head()
```

```python
# Export as a CSV
final_csv.to_csv("Resources/final_csv.csv")
```

```python
# Converted CSV file to a JSON file to match what we are used to working with 
def csv_to_json(final_csv, final_json):
    # Read CSV file
    with open(final_csv, 'r') as csvfile:
        # Assuming the first row contains the header
        reader = CsV.DictReader(csvfile)
        # Convert CSV to JSON
        data = [row for row in reader]
    # Write JSON file
    with open(final_json, 'w') as jsonfile:
        json.dump(data, jsonfile, indent=4)

# Define the path to the folder where you want to save the JSON file
folder_path = "Resources"

# Define the file name for the JSON file
json_filename = "final_json.json"

# Combine the folder path and the file name
final_json = os.path.join(folder_path, json_filename)

# Call the function with the CSV and JSON file paths
csv_to_json("Resources/final_csv.csv", final_json)
```
